```python
# We need to change its format to [Action, Advanture, Fantasy....]
# the object is a string of list so we need to convert it into a list 
# we will use ast.literal_eval to to convert the string of list to a list
import ast
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.shape
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# We need to access first 3 actor names for cast
import ast
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head()
```

```python
# We need to access the  director  from the crew
import ast
def fetchDirector(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break

  return L
```

```python
movies['crew']=movies['crew'].apply(fetchDirector)
movies.head()
```

```python
# overview is in string format so we convert it into a list for easy concatination
movies['overview'][0]
```

```python
# remove spaces inbetween words
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# remove spaces inbetween words
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# We will convert texts to vectors to find similarity score 
#The technique we use is bag of words technique of vectorization
# when vectors are formed we need to find the closest vector to our movies vector to get recommendation 
#  https://towardsdatascience.com/text-vectorization-bag-of-words-bow-441d1bfce897#:~:text=Bag%20of%20words%20is%20the,a%20word%20in%20a%20sentence.&text=The%20values%20corresponding%20to%20each,a%20word%20in%20a%20review.
```

```python
from sklearn.feature_extraction.text import  CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')           
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
# Most Frequent Words List
cv.get_feature_names()
```

```python
# Cosine distance is used where theta or the angle is used to find the similarity between movies
# distance is inversely proportinal to similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
 sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
                       
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
